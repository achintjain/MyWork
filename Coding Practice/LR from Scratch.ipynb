```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_regression
X,Y = make_regression(n_features=1, noise=20, n_samples=1000)
```

```python
plt.xlabel('Feature - X')
plt.ylabel('Target - Y')
plt.scatter(X,Y,s=5)
```

```python
mean_X = np.mean(X)
mean_Y = np.mean(Y)
```

```python
m = len(X)
```

```python
numer = 0
denom = 0
for i in range(m):
    numer += (X[i]- mean_X) * (Y[i] - mean_Y)
    denom += (X[i] - mean_X) **2
b1 = numer/denom
b0 = mean_Y - (b1 * mean_X)

print(b1,b0)
```

```python
max_X = np.max(X)
min_X = np.min(X)
x = np.linspace(min_X,max_X, 1000)
y = b0 + b1 * x

plt.scatter(X,Y,s=5)
plt.plot(x,y,c = 'r')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(X,Y)
```
