```python
import numpy as np
import pandas as pd

import seaborn as sns

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
playdata = pd.read_csv(r"C:\Users\achintj\Documents\Python Scripts\google-play-store-apps\googleplaystore.csv")
```

```python
playdata.head(50)
```

```python
gamesdata = playdata[playdata["Category"] == "GAME"]
```

```python
gamesdata.head()
```

```python
def clearstring(x):
    return x.replace("+","")
```

```python
gamesdata["Installs"] = gamesdata.Installs.apply(clearstring)
```

```python
gamesdata.head()
```

```python
def clearcomma(x):
    return int(x.replace(",",""))
```

```python
gamesdata["Installs"] = gamesdata.Installs.apply(clearcomma)
```

```python
gamesdata.sort_values(by = "Installs", ascending  = False)
```

```python
gamesdata[gamesdata.duplicated(subset = "App")].sort_values(by = "Installs", ascending  = False)
```

```python
gamesdata.drop_duplicates(subset="App", keep='first', inplace=True)
```

```python
gamesdata.sort_values(by = "Installs", ascending  = False, inplace = True)
```

```python
gamesdata.head(10)
```

```python
gamesdata.groupby(["Genres"]).count().App
```

```python
gamesdata.Genres.unique()
```

```python
plt.figure(figsize = (15,8))

plt.bar(gamesdata.Genres.unique(),gamesdata.groupby(["Genres"]).count().App)

plt.show()
```
